```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import datasets 
data = datasets.load_iris() 
#create a DataFrame 
df = pd.DataFrame(data.data, columns=data.feature_names) 
df['Target'] = pd.DataFrame(data.target) 
print(df.shape)
df.head()
# print(data.data)
# print(data.feature_names)
# print(data.target_names)
```

```python
#visualisasi hasil ConvexHull
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
import myConvexHull.CHull

plt.figure(figsize = (10, 6))
colors = ['b','r','g']
plt.title('Petal Width vs Petal Length')
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
for i in range(len(data.target_names)):
    bucket = df[df['Target'] == i]
    bucket = bucket.iloc[:,[0,1]].values
    # for j in range(len(bucket)):
        # print(bucket[j], j)
    # hull = ConvexHull(bucket) #bagian ini diganti dengan hasil implementasi ConvexHull Divide & Conquer
    hull = myConvexHull.CHull.CHull(bucket)
    plt.scatter(bucket[:, 0], bucket[:, 1], label=data.target_names[i])
    for j in range(len(bucket)):
        plt.annotate(j, (bucket[j][0], bucket[j][1]))
    for simplex in hull.simplices:
        # print(simplex)
        plt.plot(bucket[simplex, 0], bucket[simplex, 1], colors[i])
        # print(simplex[0], end=" ")
        # print(simplex[1])
        # print(bucket[simplex, 0], end="")
        # print(bucket[simplex, 1])
    print(data.target_names[i])
        
plt.legend()
```
